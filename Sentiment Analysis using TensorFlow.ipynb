## Loading Dataset

```python
from tensorflow.python.keras.datasets import imdb
(x_train, y_train),(x_test,y_test)=imdb.load_data(num_words = 10000)
```

```python
print(x_train[0])
```

```python
print(y_train[0])
```

```python
class_names = ['Negative','Positive']
```

```python
word_index = imdb.get_word_index()
print(word_index['hello'])
```

## Decoding Reveiews

```python
reverse_word_index = dict((value,key) for key, value in word_index.items())

def decode(review):
    text=''
    for i in review:
        text += reverse_word_index[i]
        text += ' '
    return text
```

```python
decode(x_train[0])
```

```python
def show_len():
    print('Len of 1st training example: ', len(x_train[0]))
    print('Len of 2nd training example: ', len(x_train[1]))
    print('Len of 1st test example: ', len(x_test[0]))
    print('Len of 2nd test example: ', len(x_test[1]))
    
show_len()
```

## Padding the examples

```python
word_index['the']
```

```python
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
```

```python
x_train = pad_sequences(x_train, value = word_index['the'], padding = 'post',maxlen = 256)
x_test = pad_sequences(x_test, value = word_index['the'], padding = 'post',maxlen = 256)
```

```python
show_len()
```

```python
decode(x_train[0])
```

## Word Embeddings

### Creating and Training the Model

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Embedding, GlobalAveragePooling1D
```

```python
model = Sequential([
    Embedding(10000, 16), #Embedding can only be used as a first layer or an input layer
    GlobalAveragePooling1D(),
    Dense(16, activation='relu'),
    Dense(1, activation = 'sigmoid')
])

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

model.summary()
```

```python
from tensorflow.python.keras.callbacks import LambdaCallback
simple_log = LambdaCallback(on_epoch_end = lambda e, l: print(e, end = '.'))

E = 20

h = model.fit(
    x_train, y_train,
    validation_split = 0.2,
    epochs = E,
    callbacks = [simple_log],
    verbose = False
)
```

### Predicitions and Evaluation

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(range(E),h.history['accuracy'],label='Training')
plt.plot(range(E),h.history['val_accuracy'],label='Validation')
plt.legend()
plt.show()
```

```python
loss, acc = model.evaluate(x_test, y_test)
print('Test set accuracy: ', acc*100)
```
